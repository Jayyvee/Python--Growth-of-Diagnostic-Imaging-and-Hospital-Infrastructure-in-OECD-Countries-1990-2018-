```python
# advanced function to search for mixed type data columns
for col in df_health.columns.tolist():
  weird = (df_health[[col]].map(type) != type(df_health[col].iloc[0])).any(axis=1)
  if len(df_health[weird]) > 0:
    print(col)
```
